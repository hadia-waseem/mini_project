importing libraries and datasets

```python
import pandas as pd
```

```python
df = pd.read_csv(r"dataset\Pokemon.csv")
df.info()
```

Total 800 values and column type 02 are secondary powers  --- indicating missing values can be ignored

```python
print(df.isnull().sum())
```

Handling repititive data and duplicate data --- no duplicate and repirtive data found

```python
print(df.duplicated().sum())
```

```python
df["Name"] = df["Name"].str.replace(r".*?\b(Mega.*)", r"\1", regex=True)
df["Name"] = df["Name"].str.replace(r"^.*?(Mega)", r"\1", regex=True).str.strip()

df.head(10)
```

```python
df.drop(columns=["#"], inplace = True)
```

```python
df["Type 2"] = df["Type 2"].fillna(df["Type 1"])
df.head(10)
```
